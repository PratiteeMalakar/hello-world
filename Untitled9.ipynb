```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Step 1: Send a request to the website
url = "https://www.ulta.com/"  # Replace with the actual URL
response = requests.get(url)

# Step 2: Parse the HTML content with BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Step 3: Extract all text from the webpage
all_text = soup.get_text()

# Step 4: Split the text into lines and clean up
lines = [line.strip() for line in all_text.splitlines() if line.strip()]

# Step 5: Convert the list of lines into a DataFrame
df = pd.DataFrame(lines, columns=["Text"])

# Step 6: Save the DataFrame to a CSV file
df.to_csv("scraped_text.csv", index=False)

print("Text data has been successfully saved to scraped_text.csv")
```

```python
import pandas as pd

# Load the three CSV files
csv1 = pd.read_csv(r"merged.csv")
csv2 = pd.read_csv(r"dataset_instagram-post-scraper_2024-08-08_21-04-49-417.csv")
csv3 = pd.read_csv("scraped_text.csv")

# Concatenate them horizontally
combined_df = pd.concat([csv1, csv2, csv3], axis=1)

# Save the combined DataFrame to a new CSV file
combined_df.to_csv("combined_data.csv", index=False)

print("The CSV files have been successfully combined into combined_data.csv")
```

```python
combined_df.head()
```

```python
# Concatenate all columns into a single string for each row
combined_df['combined_text'] = combined_df.apply(lambda row: ' '.join(row.dropna().astype(str)), axis=1)
```

```python
from langchain.embeddings import OpenAIEmbeddings

# Initialize the embeddings model
embeddings_model = OpenAIEmbeddings()

# Generate embeddings for each combined row
combined_df['embeddings'] = combined_df['combined_text'].apply(lambda text: embeddings_model.embed(text))
```
